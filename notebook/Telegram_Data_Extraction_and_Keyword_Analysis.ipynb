```python
import sys
import os
sys.path.append(os.path.abspath(os.path.join('..')))

from data_cleaning.telegram_data_extraction import TelegramDataExtractor

# Set the file path for loading the data
file_path = '../data_collection/raw_data/messages.csv'

# Set the file path for saving the filtered medical messages
output_file_path = '../data_cleaning/filtered_medical_messages.csv'

# Create an instance of TelegramDataExtractor
extractor = TelegramDataExtractor(file_path)

# Load the data
extractor.load_data()

# Display the first few rows of the dataframe
extractor.display_data()

# Clean the text
extractor.clean_text()

# Filter and display medical-related messages
medical_messages = extractor.filter_medical_keywords()
print("Filtered medical messages:", medical_messages)

# Save the filtered medical messages to a CSV file
medical_messages.to_csv(output_file_path, index=False)
print(f"Filtered medical messages saved to '{output_file_path}'")
```
